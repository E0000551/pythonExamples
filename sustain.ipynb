```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('global-data-on-sustainable-energy.csv', encoding='ISO-8859-1')
```

```python
df = df.rename(columns={
        "Entity":"Pais",
        "Year":"Año", 
        "Access to electricity (% of population)":"accessElectricity", 
        "Access to clean fuels for cooking":"cleanFuels",
        "Renewable-electricity-generating-capacity-per-capita": "renewElectricity",
        "Financial flows to developing countries (US $)": "finnancial",
        "Renewable energy share in the total final energy consumption (%)":"renewEnergyShare",
        "Electricity from fossil fuels (TWh)":"ElecFosil",
        "Electricity from nuclear (TWh)":"ElecNuclear",
        "Electricity from renewables (TWh)":"ElecRenew",
        "Low-carbon electricity (% electricity)":"LowCarbElec",
        "Primary energy consumption per capita (kWh/person)":"EnerConsPerCapita",
        "Energy intensity level of primary energy (MJ/$2017 PPP GDP)":"Intensity",
        "Value_co2_emissions_kt_by_country":"CO2",
        "Renewables (% equivalent primary energy)":"Renew",
        "gdp_growth":"gdpGrow",
        "gdp_per_capita":"gdpPerCap",
        "Density\\n(P/Km2)":"Density",
        "Land Area(Km2)":"LandArea",
        "Latitude":"Lat",
        "Longitude":"Long"
        }
    )
```
